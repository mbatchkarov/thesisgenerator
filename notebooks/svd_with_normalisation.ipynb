Check if L2-normalisation (before SVD) preserves the pairwise cosine distances between data points (it should) 

```python
%matplotlib inline
%cd '~/NetBeansProjects/thesisgenerator/'
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
from sklearn.metrics.pairwise import pairwise_distances
from scipy.spatial.distance import cosine
from numpy.testing import assert_array_almost_equal
```

```python
data = np.random.multivariate_normal([0, 0], 
                                     [[1, 0.5],[0.5, 1]],
                                     10)
colors = range(10)

plt.subplot(1,2,1)
plt.scatter(data[:,0], data[:,1], c=colors, s=250);

datan = normalize(data) # default norm is l2
plt.subplot(1,2,2)
plt.scatter(datan[:,0], datan[:,1], c=colors, s=250);

print np.linalg.norm(datan, axis=1)
assert_array_almost_equal(pairwise_distances(data, metric=cosine),
                          pairwise_distances(datan, metric=cosine))
```

Check if the same holds when doing SVD to the data

```python
from sklearn.decomposition import TruncatedSVD
colors = range(10)

data = np.random.multivariate_normal([0, 0, 0], 
                                     [[1, 0.5, 2],[3, 0.5, 1], [3, 0.5, 1]],
                                     10)

datan = normalize(data, norm='l2')

data = TruncatedSVD().fit_transform(data)
datan = TruncatedSVD().fit_transform(datan)

plt.subplot(1,2,1)
plt.scatter(data[:,0], data[:,1], c=colors, s=250);

plt.subplot(1,2,2)
plt.scatter(datan[:,0], datan[:,1], c=colors, s=250);

assert_array_almost_equal(pairwise_distances(data, metric=cosine), 
                          pairwise_distances(datan, metric=cosine))
```

Normalisation does *not* seem to have an effect on the pairwise cosine similarities of data points even after SVD.
