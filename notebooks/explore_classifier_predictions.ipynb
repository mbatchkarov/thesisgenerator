```python
def get_full_test_matrix(filtered_matrix, row_mask):
    '''
    Reconstructs a full matrix from rows that survived when the row filter `mask` was applied.
    Puts the remaining rows where they were before the mask was applied, the rest is np.nan
    
    :type filtered_matrix: csr_matrix
    '''
    full_mat = np.zeros((len(row_mask), filtered_matrix.shape[1]))
    full_mat.fill(np.nan)
    new_idx = 0 # index into the filtered array
    for old_idx, selected in enumerate(row_mask):
        if selected:
            # must have been in the old array, copy it over
            tmp = filtered_matrix[new_idx, :].A.ravel()
            full_mat[old_idx, :] = tmp
            new_idx += 1

    assert_array_equal(filtered_matrix.A, full_mat[row_mask, :])
    return full_mat

# do a quick check of the function above 
unit_test_mask = np.array([True, False, True], dtype=np.bool)
unit_test_matrix = csr_matrix(np.array([[0, 0, 0], [1, 1, 1], [2, 2, 2]]))
unit_test_recovered_matrix = np.array([[0, 0, 0], [np.nan]*3, [2, 2, 2]])
assert_array_equal(unit_test_recovered_matrix,
                   get_full_test_matrix(unit_test_matrix[unit_test_mask, :], unit_test_mask))

full_add_matrix = get_full_test_matrix(add.test_matrix, add.test_mask)
full_socher_matrix = get_full_test_matrix(socher.test_matrix, socher.test_mask)
full_baseline_matrix = get_full_test_matrix(baseline.test_matrix, baseline.test_mask)
```

Log odds
=============
Calculate the log odds of each feature $f_i$ as 

\begin{align} LO(f_i)=ln\frac{P(f_i|c_0)}{P(f_i|c_1)}  \end{align}

```python
log_odds = calculate_log_odds(baseline.tr_matrix, baseline.y_tr)
lo_dict = {feat.tokens_as_str():log_odds[idx] for idx, feat in baseline.inv_voc.items()}
```

```python
def pprint_matrix(y, inv_voc, doc_vectors, doc_ids, log_odds):
    all_columns = [inv_voc[foo].tokens_as_str() for foo in range(len(inv_voc))]
    df = DataFrame(doc_vectors, columns=all_columns, index=doc_ids)
    
    THRESH = 5
    for id, feats in sorted(df.T.to_dict().items(), key=itemgetter(0)): # walk docs in order of id
        print 'Document %d (true class=%d):'%(id, y[id])
        approx_sum_log_odds = 0
        for feat, doc_count in sorted(feats.items(), key=itemgetter(1), reverse=True): # term counts in doc in order of decreasing freq
            if not doc_count:
                continue
            print '\t%s, count %d, LO %.2f'%(feat, doc_count, log_odds[feat])
            if log_odds[feat] > THRESH:
                approx_sum_log_odds += THRESH * doc_count
            elif log_odds[feat] < -THRESH:
                approx_sum_log_odds += - (THRESH * doc_count)
            else:
                approx_sum_log_odds += log_odds[feat] * doc_count
        print '\t\t-----Approx sum of log odds: %2.2f'%approx_sum_log_odds


# pprint_matrix(targets, baseline.inv_voc, interesting_doc_vectors_baseline, interesting_doc_ids, lo_dict)
# print '-----------------------------------------------------'
# pprint_matrix(targets, socher.inv_voc, interesting_doc_vectors_socher, interesting_doc_ids, lo_dict)
# print '-----------------------------------------------------'
# pprint_matrix(targets, add.inv_voc, interesting_doc_vectors_add, interesting_doc_ids, lo_dict)
```

```python
class Replacement(object):
    """
    This class holds information about replacements that occurs at decode time, such as
    the classificational log odds score of the replacement,
    its thesaurus similarity to the original, the number of times this replacement was made, etc
    """

    def __init__(self, replacement, lo_replacement, unlab_freq_replacement, sim, rank):
        self.replacement = replacement
        self.lo_replacement = lo_replacement
        self.unlab_freq_replacement = unlab_freq_replacement
        self.sim = sim
        self.rank = rank
    
    def __str__(self):
        return '%s (LO %2.1f, sim %2.2f, rank %d, freq %d)'%(self.replacement, self.lo_replacement,
                                                             self.sim, self.rank,
                                                             self.unlab_freq_replacement)
    def __repr__(self):
        return str(self)
```

```python
# do analysis at the document level- for each doc, find the replacements that are being made
import pandas as pd

def get_replacements_df(paraphrases_file, lo_dict, feat_freq):
    # this bit lifted from  signifier_internals_analysis.get_replacements_df  
    df = pd.read_csv(paraphrases_file, sep=', ')
    counts = df.groupby('feature').count()
    df = df.drop_duplicates()
    df.set_index('feature', inplace=True)
    df['count'] = counts.available_replacements
    
    # this bit lifted from signifier_internals_analysis.analyse_replacement_ranks_and_sims
    res = defaultdict(list)
    for i in range(1, 4):
        replacements = df['replacement%d' % i]
        sims = df['replacement%d_sim' % i]
        ranks = df['replacement%d_rank' % i]
        for original, replacement, sim, rank in zip(df.index, replacements, sims, ranks):
            if (sim > 0) and (not np.isnan(sim)) and (not np.isnan(rank)):  # filter out NaN-s
                res[original].append(Replacement(replacement, lo_dict[replacement], feat_freq[replacement], sim, rank))
    return res

socher_replacements = get_replacements_df('statistics/stats-exp%d-0-cv0-ev.par.csv'%SOCHER_NUM, lo_dict, feat_freq)
add_replacements = get_replacements_df('statistics/stats-exp%d-0-cv0-ev.par.csv'%ADD_NUM, lo_dict, feat_freq)
print socher_replacements['open/J_market/N']
print
print add_replacements['open/J_market/N']

# freeze the dicts
# socher_replacements.default_factory = None
# add_replacements.default_factory = None
```

Gigaword frequency of replacements
===============

Find features that both Socher and Add had replacements for, compare the frequency of these replacements in Gigaword

```python
socher_ranks, socher_freqs, socher_sims  = [], [], []
add_ranks, add_freqs, add_sims = [], [], []
for feat in socher_replacements.keys():
    if (not add_replacements[feat]) or (not socher_replacements[feat]):
        # these are both defaultdicts, and the cell above queries them for keys they may not contain
        # this check skips over the values that get inserted this way and prevent np.mean([])        
        continue
    socher_ranks.append(np.mean([r.rank for r in socher_replacements[feat]]))
    socher_freqs.append(np.mean([r.unlab_freq_replacement for r in socher_replacements[feat]]))
    socher_sims.append(np.mean([r.sim for r in socher_replacements[feat]]))
    add_ranks.append(np.mean([r.rank for r in add_replacements[feat]]))
    add_freqs.append(np.mean([r.unlab_freq_replacement for r in add_replacements[feat]]))
    add_sims.append(np.mean([r.sim for r in add_replacements[feat]]))

figsize(12,8)
f, axarr = plt.subplots(2, 2, sharey=True);
axarr[0, 0].hist(socher_ranks, bins=20); 
axarr[0, 0].set_title('Socher ranks')
axarr[0, 1].hist(add_ranks, bins=20);
axarr[0, 1].set_title('Add ranks')
axarr[1, 0].hist(socher_freqs, bins=20);
axarr[1, 0].set_title('Socher freqs')
axarr[1, 1].hist(add_freqs, bins=20);
axarr[1, 1].set_title('Add freqs')

# hide x ticks for top plots and y ticks for right plots
plt.setp([a.get_xticklabels() for a in axarr.ravel()], visible=True, rotation=-30);
# plt.setp([a.get_yticklabels() for a in axarr.ravel()], visible=False);
plt.tight_layout()
```

Analysis
===========
Socher tends to find replacements with a much lower frequency (in gigaword), probably due to normalisation.

The difference is significant


```python
from scipy.stats import ttest_ind, ttest_rel
print 'Mean gigaw freq of Socher and Add replacements: %2.1f, %2.1f'%(mean(socher_freqs), mean(add_freqs))
print 'T-test p-value based on %d samples: %2.2f'%(len(socher_freqs), ttest_ind(socher_freqs, add_freqs)[1])
```

However, the normalisation Socher does is non-trivial and different to the commonly used L2-normalisation (e.g. Vecchi et al, Polajnar and Clark). Because we use cosine similarity, standard L2-normalisation does **not** result in different neighbour sets. The same holds when performing SVD prior to nearest-neighbour search.

```python
from scipy.stats import pearsonr
assert len(add_sims) == len(add_freqs)
assert len(socher_sims) == len(socher_freqs)
plt.subplot(1,2,1)
plt.scatter(add_sims, np.sqrt(add_freqs));
corr = pearsonr(add_sims, np.sqrt(add_freqs))[0]
plt.ylabel('Log (gigaw freq)');
plt.title('Add, corr=%.2f'%corr);
           
plt.subplot(1,2,2)
plt.scatter(socher_sims, np.sqrt(socher_freqs));
corr = pearsonr(socher_sims, np.sqrt(socher_freqs))[0]
plt.title('Socher, corr=%.2f'%corr);
```

It doesn't seem to be the case that frequency in Gigaword is related to similarity.

However, the similarity scores of the Add composer have very little range, which is a known problem. This is particularly pronounced when using window features. This is worth looking into.
