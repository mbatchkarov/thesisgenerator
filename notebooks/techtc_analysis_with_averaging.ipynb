```python
%matplotlib inline
import matplotlib.pyplot as plt
from itertools import chain
import pandas as pd
import numpy as np
from statsmodels.stats.multicomp import MultiComparison
import statsmodels.api as sm
from critical_difference.plot import do_plot, print_figure
from collections import defaultdict
```

```python
cols = ['num', 'unlabelled', 'labelled', 'svd', 'composer', 'document_features', 
        'thesaurus_features', 'baronified', 'use_similarity', 'use_random_neighbours', 
        'decode_handler']
df = pd.read_csv('../../ExpLosion/experiments.txt', header=0, names=cols, sep=',', index_col=0)
```

```python
CV_FOLDS = 25
METRIC = 'accuracy_score'

def get_composer_name(df, n):
    return df.composer[n]

def _get_cv_scores_single_experiment(df, n, classifier):
    # human-readable name
    composer_name = get_composer_name(df, n)
    # get scores for each CV run- these aren't in the database
    # only at size 500
    outfile = '../conf/exp{0}/output/exp{0}-0.out-raw.csv'.format(n)
    scores = pd.read_csv(outfile)
    mask = (scores.classifier == classifier) & (scores.metric == METRIC)
    ordered_scores = scores.score[mask].tolist()
    return ordered_scores, '%d-%s' % (n, composer_name)
    
def get_scores(df, exp_ids, classifier='MultinomialNB'):
    # get human-readable labels for the table
    data = []
    composers = []

    print('Running significance for experiments %r' % exp_ids)
    for n in exp_ids:
        for something in ([n] if isinstance(n, int) else n):
            scores, composer = _get_cv_scores_single_experiment(df, something, classifier)
            data.extend(scores)
            composers.extend([composer]*CV_FOLDS)

    return data, composers

def run_significance(data, composers):
    mod = MultiComparison(data, composers, group_order=sorted(set(composers)))
    a = mod.tukeyhsd(alpha=0.01)
    # reject hypothesis that mean is the same? rej=true means a sign. difference exists

    '''
    a looks like this:

    Multiple Comparison of Means - Tukey HSD,FWER=0.01
    ===============================================
     group1  group2 meandiff  lower   upper  reject
    -----------------------------------------------
    57-APDT 58-APDT -0.0487  -0.0759 -0.0215  True
    59-APDT 60-APDT -0.0124  -0.0395  0.0148 False
    -----------------------------------------------
    '''
    data = str(a).split('\n')
    desc = data[0]
    header = data[2].split()
    header[-1] = 'significant'
    rows = [row.split() for row in data[4:-1]]
    return pd.DataFrame(rows, columns=header)

def get_demsar_diagram(df, names, scores):
    idx = np.argsort(scores)
    scores = list(scores[idx])
    names = list(names[idx])

    def get_insignificant_pairs(*args):
        mylist = []
        for a, b, significant_diff in zip(df.group1, df.group2, df.significant):
            if significant_diff == 'False':  # convert str
                mylist.append((names.index(a), names.index(b)))
        return sorted(set(tuple(sorted(x)) for x in mylist))

    fig = do_plot(scores, get_insignificant_pairs, names)
    fig.set_canvas(plt.gcf().canvas)
    print_figure(fig, "%s.png" % ('_'.join(sorted(set(names)))), format='png')
```

```python
# plot critical difference diagram given a list of experiment numbers
all_scores, all_names = get_scores(df, [11, 10, [12,13,14]]) # all experiments that are grouped must only differ by dataset
sign_table = run_significance(np.array(all_scores), all_names)
mean_scores = pd.DataFrame(np.array([all_names, all_scores]).T, 
                           columns = ['name', 'score']).convert_objects(convert_numeric=True)
mean_scores = mean_scores.groupby('name').mean().reset_index()
fig = get_demsar_diagram(sign_table, mean_scores.name, mean_scores.score*100)
```
