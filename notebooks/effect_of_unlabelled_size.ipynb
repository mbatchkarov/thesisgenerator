Let's investigate if the same holds for the document classification task. For simiplicity, we only look at `word2vec` vectors with simple composion on the Reuters data set.

```python
%cd ~/NetBeansProjects/thesisgenerator/
%matplotlib inline
%config InlineBackend.figure_format = 'svg' 
import os
import peewee as pw
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from thesisgenerator.utils.db import ClassificationExperiment, Vectors
```

```python
def get_exp_ids_for_varying_amounts_of_gigaword():
    ids = []
    query = (ClassificationExperiment
             .select(ClassificationExperiment.id)
             .join(Vectors)
             .where(ClassificationExperiment.decode_handler == 'SignifiedOnlyFeatureHandler',
                    ClassificationExperiment.document_features == 'AN_NN',
                    ClassificationExperiment.labelled == 'reuters21578/r8-tagged-grouped',
                    ClassificationExperiment.use_similarity == 0,
                    Vectors.algorithm == 'word2vec',
                    Vectors.rep == 0).tuples())
    for f in query:
        ids.append(f[0])
    return set(ids)

def get_exp_ids_for_reruns_of_word2vec():
    ids = []
    query = (ClassificationExperiment
             .select(ClassificationExperiment.id)
             .join(Vectors)
             .where(ClassificationExperiment.decode_handler == 'SignifiedOnlyFeatureHandler',
                    ClassificationExperiment.document_features == 'AN_NN',
                    ClassificationExperiment.labelled == 'reuters21578/r8-tagged-grouped',
                    ClassificationExperiment.use_similarity == 0,
                    Vectors.algorithm == 'word2vec',
                    Vectors.unlabelled_percentage == 100).tuples())
    for f in query:
        ids.append(f[0])
    return set(ids)
```

```python
# based on code lifted from ExpLosion
METRIC_DB = 'macrof1'
METRIC_CSV_FILE = 'macroavg_f1'

def get_vectors_field(n, field_name):
    vectors = ClassificationExperiment.get(id=n).vectors
    return getattr(vectors, field_name) if vectors else None

def get_cv_scores_single_experiment(n, classifier):
    composer_name = get_vectors_field(n, 'composer')
    outfile = '../thesisgenerator/conf/exp{0}/output/exp{0}-0.out-raw.csv'.format(n)
    if not os.path.exists(outfile):
        print('Output file for exp %d missing' % n)
        return None, composer_name, None
    scores = pd.read_csv(outfile)
    mask = (scores.classifier == classifier) & (scores.metric == METRIC_CSV_FILE)
    ordered_scores = scores.score[mask].tolist()
    return ordered_scores, '%d-%s' % (n, composer_name), get_vectors_field(n, 'rep')


def get_scores(exp_ids, classifier='MultinomialNB', cv_folds=25):
    # get human-readable labels for the table
    data = []
    composers = []
    folds = []
    percent = []
    reps = []
    

    for exp_number in exp_ids:
        perc = ClassificationExperiment.get(id=exp_number).vectors.unlabelled_percentage
        composer = get_vectors_field(exp_number, 'composer')
        scores, wtf, rep_no = get_cv_scores_single_experiment(exp_number, classifier)
        if scores:
            composers.extend([composer] * cv_folds)
            percent.extend([perc] * cv_folds)
            reps.extend([rep_no] * cv_folds)
            folds.extend('cv%d' % foo for foo in range(cv_folds))
            data.extend(scores)
    return data, composers, folds, percent, reps
```

```python
ids = get_exp_ids_for_varying_amounts_of_gigaword()
print (ids)
data, composers, folds, percent, reps = get_scores(ids)
df1 = pd.DataFrame({'F1':data, 
                   'Composer':composers,
                   'cvfold':folds,
                   '% of Gigaword':percent})
```

# Overlap of neighbours for $word_i$ across `word2vec` runs
Dice coefficient over a small sample, neighbours selected out of the entire `word2vec` vocabulary

```python
def dice(n1, n2):
    return 2 * len(set(n1) & set(n2))/ (len(n1) + len(n2))

for t in thes:
    t.init_sims()
pairs = []
sims = []
s = sample([x for x in thes[0].keys() if x[-1] in 'NJ'], 50)
for i,j in combinations(range(len(thes)), 2):
    print('Doing pair %d and %d'%(i,j))
    for num, foo in enumerate(s):
        n1 = [x[0] for x in thes[i].get_nearest_neighbours(foo)]
        n2 = [x[0] for x in thes[j].get_nearest_neighbours(foo)]
        sims.append(dice(n1, n2))
    pairs.extend(['%d-%d'%(i,j)]*(len(sims) - len(pairs)))
```

```python
df3 = pd.DataFrame({'Pair': pairs,
                   'Dice': sims,
                   'Words' : s * 3})
g = sns.FacetGrid(df3, col="Pair");
g.map(sns.distplot, 'Dice');
```
