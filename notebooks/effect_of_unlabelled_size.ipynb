```python
%cd ~/NetBeansProjects/thesisgenerator/
%matplotlib inline
import os
import peewee as pw
import pandas as pd
import seaborn as sns
import numpy as np
from thesisgenerator.utils.db import ClassificationExperiment, Results, Vectors
```

```python
ids = []
query = (Results
         .select(ClassificationExperiment.id)
         .join(ClassificationExperiment)
         .join(Vectors)
         .where(ClassificationExperiment.decode_handler == 'SignifiedOnlyFeatureHandler',
                ClassificationExperiment.document_features == 'AN_NN',
                ClassificationExperiment.labelled == 'reuters21578/r8-tagged-grouped',
                ClassificationExperiment.use_similarity == 0,
                Vectors.algorithm == 'word2vec').tuples())
for f in query:
    ids.append(f[0])
ids=set(ids)
```

```python
# lifted from ExpLosion
METRIC_DB = 'macrof1'
METRIC_CSV_FILE = 'macroavg_f1'

def get_composer_name(n):
    vectors = ClassificationExperiment.get(id=n).vectors
    return vectors.composer if vectors else None

def _get_cv_scores_single_experiment(n, classifier):
    # human-readable name
    composer_name = get_composer_name(n)
    # get scores for each CV run- these aren't in the database
    # only at size 500
    outfile = '../thesisgenerator/conf/exp{0}/output/exp{0}-0.out-raw.csv'.format(n)
    if not os.path.exists(outfile):
        print('Output file for exp %d missing' % n)
        return None, composer_name
    scores = pd.read_csv(outfile)
    mask = (scores.classifier == classifier) & (scores.metric == METRIC_CSV_FILE)
    ordered_scores = scores.score[mask].tolist()
    return ordered_scores, '%d-%s' % (n, composer_name)


def get_scores(exp_ids, classifier='MultinomialNB', cv_folds=25):
    # get human-readable labels for the table
    data = []
    composers = []
    folds = []
    percent = []

    for exp_number in exp_ids:
        perc = ClassificationExperiment.get(id=exp_number).vectors.unlabelled_percentage
        composer = get_composer_name(exp_number)
        scores, _ = _get_cv_scores_single_experiment(exp_number, classifier)
        if scores:
            composers.extend([composer] * cv_folds)
            percent.extend([perc] * cv_folds)
            folds.extend('cv%d'%foo for foo in range(cv_folds))
            data.extend(scores)
    return data, composers, folds, percent

data, composers, folds, percent = get_scores(ids)
```

```python
df = pd.DataFrame({'F1':data, 
                   'Composer':composers,
                   'cvfold':folds,
                   '% of Gigaword':percent})
sns.tsplot(df, time='% of Gigaword', condition='Composer', value='F1', unit='cvfold');
```
